```python
import glob 
import os 
import numpy as np 
from jupyterthemes import jtplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
home = os.path.join(os.path.expanduser('~'))
data_path = (home + '/Dropbox/loki_0.5/analysis/pupil/processed_data/pupil_features/')
fig_path = (home + '/Dropbox/loki_0.5/figures/')
```

```python
data_path
```

```python
feature_fn = ('*feature_df.csv')
feature_files = glob.glob(os.path.join(data_path, feature_fn))
# assert len(feature_files) == 9, 'check feature files'
# feature_dfs_pc0 = [pd.read_csv(file) for file in feature_files]

feature_df = pd.read_csv(feature_files[0]).drop(columns=['subj_id', 'reward_code',
                                                        'trial'])
labels = feature_df.columns
n_features = len(labels)
```

```python
feature_fn = ('*_pc0_feature_importance.csv')
feature_files = glob.glob(os.path.join(data_path, feature_fn))
# assert len(feature_files) == 9, 'check feature files'
feature_dfs_pc0 = [pd.read_csv(file) for file in feature_files]

feature_imp_df_pc0 = pd.concat(feature_dfs_pc0)
```

```python
feature_imp_df_pc0.subj_id
```

```python
feature_fn = ('*_pc1_feature_importance.csv')
feature_files = glob.glob(os.path.join(data_path, feature_fn))
# assert len(feature_files) == 9, 'check feature files'
feature_dfs_pc1 = [pd.read_csv(file) for file in feature_files]

feature_imp_df_pc1 = pd.concat(feature_dfs_pc1)
```

```python
feature_fn = ('*_pc2_feature_importance.csv')
feature_files = glob.glob(os.path.join(data_path, feature_fn))
# assert len(feature_files) == 9, 'check feature files'
feature_dfs_pc2 = [pd.read_csv(file) for file in feature_files]

feature_imp_df_pc2 = pd.concat(feature_dfs_pc2)
```

```python
feature_imp_df_pc0
```

```python
labels
formatted_labels = ['peak amp.', 'peak width', 'mean diameter', 
                   'peak onset', 'peak offset', 
                    'peak amp. latency', 'auc']
```

```python
sns.barplot("feature", "abs_pca_comp", data=feature_imp_df_pc0, color='gray')
plt.xticks(np.arange(0, n_features), labels=formatted_labels, fontsize=12, 
                  rotation=60)
plt.ylabel('Feature importance')
plt.title('Prinicipal Component 1: Timing')
plt.xlabel('')
```

```python
sns.barplot("feature", "abs_pca_comp", data=feature_imp_df_pc1, color='gray')
plt.xticks(np.arange(0, n_features), labels=formatted_labels, fontsize=12, 
                  rotation=60)
plt.ylabel('Feature importance')
plt.title('Prinicipal Component 2: Mass')
plt.xlabel('')
```

```python
sns.barplot("feature", "abs_pca_comp", data=feature_imp_df_pc2, color='gray')
plt.xticks(np.arange(0, n_features), labels=formatted_labels, fontsize=12, 
                  rotation=60)
plt.ylabel('Feature importance')
plt.title('Prinicipal Component 3')
plt.xlabel('')
```

```python
sns.set(font_scale=1.5, style='whitegrid')  

g = sns.barplot("feature", "abs_pca_comp", data=feature_imp_df_pc0, 
            hue='subj_id', palette=sns.color_palette("Paired",4))
plt.xticks(np.arange(0, n_features), labels=formatted_labels, fontsize=13, 
                  rotation=60)
plt.ylabel('Feature importance')
plt.title('Principal Component 1: Timing')
plt.xlabel('')
g.legend(loc='center right', bbox_to_anchor=(1.5, 0.5), ncol=1)
legend = g.legend_

for t, l in zip(legend.texts,("S1", "S2", "S3", "S4")):
    t.set_text(l)
    
plt.savefig(os.path.join(fig_path, 'feature_imp_pc1_timing.png'), dpi=600, bbox_inches='tight', transparent=0)
```

```python
sns.set(font_scale=1.5, style='whitegrid')  

g = sns.barplot("feature", "abs_pca_comp", data=feature_imp_df_pc1, 
            hue='subj_id', palette=sns.color_palette("Paired",4))
plt.xticks(np.arange(0, n_features), labels=formatted_labels, fontsize=13, 
                  rotation=60)
plt.ylabel('Feature importance')
plt.title('Principal Component 2: Mass')
plt.xlabel('')
g.legend(loc='center right', bbox_to_anchor=(1.5, 0.5), ncol=1)
legend = g.legend_

for t, l in zip(legend.texts,("S1", "S2", "S3", "S4")):
    t.set_text(l)

plt.savefig(os.path.join(fig_path, 'feature_imp_pc2_mass.png'), dpi=600, bbox_inches='tight', transparent=0)
```
